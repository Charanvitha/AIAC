```python
def is_armstrong(number):
    """
    Checks whether a number is an Armstrong number.
    """

    digits = str(number)          # Convert number to string
    power = len(digits)           # Count digits

    total = 0

    for digit in digits:          # Loop through each digit
        total += int(digit) ** power

    return total == number        # Compare sum with original
```

```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n - i - 1):

            # Swap if elements are in wrong order
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]

    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)
```

```python
products = {
    "phone": ["case", "screen protector", "charger"],
    "laptop": ["mouse", "keyboard", "USB hub"],
    "camera": ["tripod", "memory card"]
}

def recommend(item):
    if item not in products:
        return [], "No data available."

    recs = products[item]

    explanation = f"Because you viewed '{item}', users often buy: {', '.join(recs)}."

    return recs, explanation


item = input("Enter product name: ")

recommendations, reason = recommend(item)

print("Recommended:", recommendations)
print("Reason:", reason)
```
