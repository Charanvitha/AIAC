```python
def fibonacci(n):
    """
    Generates Fibonacci numbers up to n terms.
    """

    prev, curr = 0, 1

    for _ in range(n):
        print(prev, end=" ")
        prev, curr = curr, prev + curr


n = int(input("Enter number of terms: "))
fibonacci(n)
```

```python

def fib_iter(n):
    prev, curr = 0, 1

    for _ in range(n):
        print(prev, end=" ")
        prev, curr = curr, prev + curr
```

```python
def fib_rec(n):
    if n <= 1:
        return n
    return fib_rec(n - 1) + fib_rec(n - 2)


n = int(input("Enter number of terms: "))

for i in range(n):
    print(fib_rec(i), end=" ")
```
