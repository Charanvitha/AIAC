Task 1: AI-Generated Logic Without Modularization (No Functions)
Prompt Given to AI
Generate a Python program to calculate factorial using loops only.
Do not use functions.

```python
n = int(input("Enter a number: "))
result = 1

for i in range(1, n + 1):
    result = result * i

print("Factorial is:", result)
```

Task 2: AI Code Optimization & Cleanup
Prompt Given to AI
Optimize this code by reducing variables and improving readability.

```python
n = int(input("Enter a number: "))
factorial = 1

for i in range(2, n + 1):
    factorial *= i

print("Factorial is:", factorial)
```

Task 3: Modular Design Using Functions
Prompt Given to AI
Create a factorial program using a function with clear variable names and comments

```python
def calculate_factorial(number):
    result = 1
    for i in range(1, number + 1):
        result *= i
    return result

num = int(input("Enter a number: "))
print("Factorial is:", calculate_factorial(num))
```

Task 5: Iterative vs Recursive AI-Generated Code
Prompt Given to AI
Generate factorial using iteration and recursion.

```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)

print(factorial_recursive(5))
```
